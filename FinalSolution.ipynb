### Defining functions for data analysis

    - filling missing info - This function replaces ? with nan
    - variable feature - Appends the object data structures from a dataframe into a list and returns it
    - var_numeric - Applies Label Encoding and transforms the variables
    - Reshape data - Using encoding and using minmax scaling to reshape the data so that it can be split for the model
    - plot curve - Function to plot the roc and PR

```python
def fill_missing_info(data):
    """
    
    Replacing ? with nan in the dataset
    
    """
    data = data.replace(" ?", np.nan)
    
    for column in missing_col:
        top = data[column].describe().top
        data[column] = data[column].fillna(value=top)
        
    return data
```

```python
def get_obj(dataframe):
    """
    
    Variable Features
    
    """
    get_objs = []
    for column in dataframe.columns:
        if dataframe[column].dtype == "object":
            get_objs.append(column)

    return get_objs
```

```python
def var_numeric(dataframe):
   
    get_vars = dataframe.columns
    le = sklearn.preprocessing.LabelEncoder()
    
    for col in get_vars:
        dataframe[col] = le.fit_transform(dataframe[col])
        print("{col} transform to {num_col}".format(col=col,num_col='num_'+col))

    return dataframe
```

```python
def reshape_data(dataframe,encoder = "label"):
    
    # using encoding for variables
    if encoder == "label":
        df_trans = var_numeric(dataframe)
    else:
        var_feature = get_obj(dataframe)
        df_num = dataframe.drop(var_feature,axis=1)
        df_var = dataframe[var_feature]

        df_trans = pd.get_dummies(df_var)
        
        for col in list(df_num.columns):
            df_trans[col] = sklearn.preprocessing.minmax_scale(df_num[col])
    
    return df_trans
```

```python
def plot_Curve(dic,y_test):
    plt.figure(figsize=(20,20))
    for key in dic.keys():
        pred=dic.get(key)
        # plot PR Curve
        plt.subplot(211)
        pr, re, _ = sklearn.metrics.precision_recall_curve(y_test,pred)
        plt.plot(re,pr,label=key,linewidth = 3)
        plt.legend(loc='best',prop = {"size":20})
        plt.xlabel('Recall',fontsize ='20' )
        plt.ylabel('Precision',fontsize= '20' )
        plt.title("PR_Curve")
        
        #plot ROC Curve
        plt.subplot(212)
        fpr, tpr, _  = sklearn.metrics.roc_curve(y_test,pred)
        plt.plot(fpr,tpr,label=key,linewidth = 3)
        plt.legend(loc='best',prop = {"size":20})
        plt.xlabel('False Positive Rate',fontsize ='20')
        plt.ylabel('True Positive Rate',fontsize= '20' )
        plt.title("ROC_Curve")
```

### Feature Engineering

    - Adding new features
    - Tranforming features
    - Dropping some feaatures

```python

# medicine name
medicines = df.columns[24:-3]
df=fill_missing_info(df)

# count the sum of all reactions
df["num_no"] = np.sum(df[medicines]=="No",axis=1)
df["num_down"] = np.sum(df[medicines]=="Down",axis=1)
df["num_steady"] = np.sum(df[medicines]=="Steady",axis=1)
df["num_up"] = np.sum(df[medicines]=="Up",axis=1)


# count revisits of all patients in this table
df["revisits"]=df["patient_nbr"].map(df["patient_nbr"].value_counts())

# count how many times a patient came in a year
df["revisits_yearly"]=df[['number_outpatient', 'number_emergency', 'number_inpatient']].sum(axis=1)
```

```python
# transfer labels to numeric values
df["readmitted"] = sklearn.preprocessing.LabelEncoder().fit_transform(df["readmitted"])

# transform feature "weight" so that it can be used for the model

df.loc[:,"weight"] = df["weight"].replace("?",0)
df.loc[:,"weight"] = df["weight"].replace("\>|\[|\-.*","",regex=True).astype("float")

# reclassify the discarge dispostion type into bigger category
df.loc[:,"age"] = df["age"].replace("\[|\-.*","",regex=True).astype("int64")

# These are unknown values

# NULL, Not Mapped, Unknown/Valid
df.loc[:,"discharge_disposition_id"]=df["discharge_disposition_id"].replace([18,25,26],0)

# Discharged with assistance/ or without
df.loc[:,"discharge_disposition_id"]=df["discharge_disposition_id"].replace([1,6,8,13,14],1) 

# Discharged/traansferred
df.loc[:,"discharge_disposition_id"]=df["discharge_disposition_id"].replace([2,3,4,5],2) 

# other
df.loc[:,"discharge_disposition_id"]=df["discharge_disposition_id"].replace([9,10,12,13,14,15,16,17,22,23,24],3)

# death
df.loc[:,"discharge_disposition_id"]=df["discharge_disposition_id"].replace([11,19,20,21],4) 

# Change diag code to bigger classfication
# based on the standard of ICD-9 format

dignose_col = ["diag_1","diag_2","diag_3"]
diag_code = [-1,0,139,239,279,289,319,389,459,519,579,629,679,709,739,759,779,799,999,1999,2001]
df.loc[:,dignose_col]=df[dignose_col].replace("^\?",0,regex=True)
df.loc[:,dignose_col]=df[dignose_col].replace("^V",1000,regex=True)
df.loc[:,dignose_col]=df[dignose_col].replace("^E",2000,regex=True)

df[dignose_col].astype("float64")
for i in dignose_col:
    df.loc[:,i] = pd.cut(pd.to_numeric(df[i]),diag_code,labels = False)
```

```python
# dropping some features

ratio = df[df["readmitted"]==0].shape[0]/df[df["readmitted"]==1].shape[0]
label = df["readmitted"]
features = df.drop(["readmitted","encounter_id","patient_nbr"],axis=1)
features = features.drop(medicines,axis=1)
```
